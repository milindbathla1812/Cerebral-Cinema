```python
import h5py
import numpy as np
from scipy.signal import detrend
from scipy.stats import zscore
import os
```

```python
h5_file = r"D:\algonauts_2025.competitors\fmri\sub-03\func\sub-03_task-friends_space-MNI152NLin2009cAsym_atlas-Schaefer18_parcel-1000Par7Net_desc-s123456_bold.h5"
with h5py.File(h5_file, "r") as f:
    def print_structure(name, obj):
        print(name)

    f.visititems(print_structure)
```

```python

def preprocess_fmri(data, tr=1.49, lag_sec=5):
    """
    data: (T, parcels)

    Returns:
        preprocessed data
    """

    # 1. Detrend
    data = detrend(data, axis=0)

    # 2. Z-score
    data = zscore(data, axis=0)

    # 3. HRF lag correction
    lag_trs = int(round(lag_sec / tr))

    if lag_trs > 0:
        data = data[lag_trs:]

    return data.astype(np.float32)
```

```python
h5_file = r"D:\algonauts_2025.competitors\fmri\sub-03\func\sub-03_task-friends_space-MNI152NLin2009cAsym_atlas-Schaefer18_parcel-1000Par7Net_desc-s123456_bold.h5"
save_dir = r"D:\algonauts_2025.competitors\preprocessed_sub3_npy"

os.makedirs(save_dir, exist_ok=True)

with h5py.File(h5_file, "r") as f:

    for episode in f.keys():

        data = f[episode][:]

        data = preprocess_fmri(
            data,
            tr=1.49,
            lag_sec=5
        )

        np.save(
            os.path.join(save_dir, f"{episode}.npy"),
            data
        )

        print(
            episode,
            "saved",
            data.shape
        )
```
